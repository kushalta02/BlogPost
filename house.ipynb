```python
import pandas as pd
# from google.colab import files
# uploaded=files.upload()
df=pd.read_csv('House_Rent_Dataset.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Rent"].fillna(df["Rent"].median(), inplace=True)
```

```python
# df.isnull().sum()
df['Area Locality'].fillna(df["Area Locality"].mode(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Bathroom'].fillna(df['Bathroom'].median(),inplace=True)
df['Area Locality'].fillna(df['Area Locality'].mode(),inplace=True)
df['Furnishing Status'].fillna(df['Furnishing Status'].mode(),inplace=True)

```

```python
df.isnull().sum()
```

```python
# df['Bathroom'].fillna(df['Bathroom'].median(),inplace=True)
df['Area Locality'].fillna(df['Area Locality'].mode(),inplace=True)
df['Furnishing Status'].fillna(df['Furnishing Status'].mode(),inplace=True)
```

```python
df.isnull().sum()
```

```python
# Bathroom (numeric)
df['Bathroom'] = df['Bathroom'].fillna(df['Bathroom'].median())

# Furnishing Status (categorical)
df['Furnishing Status'] = df['Furnishing Status'].fillna(df['Furnishing Status'].mode()[0])

# Area Locality (categorical)
df['Area Locality'] = df['Area Locality'].fillna(df['Area Locality'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```
